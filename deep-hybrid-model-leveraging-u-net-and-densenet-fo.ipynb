**What is Pneumonia?**

**Pneumonia represents an infection in a single or each lungs that may be resulting from bacteria, viruses, or fungi. The sickness in most cases affects the alveoli - the tiny air sacs where fuel trade takes place - causing them to turn out to be inflamed and fill with fluid or purulent cloth. This inflammatory manner significantly impairs the lung's ability to switch oxygen to the bloodstream.**

**The severity of pneumonia varies considerably depending on numerous factors together with the causative organism, the affected person's age, and ordinary fitness status. at the same time as some instances present as mild "taking walks pneumonia" that slightly requires mattress rest, others may be severe enough to necessitate hospitalization**


![Pneumonia](https://my.clevelandclinic.org/-/scassets/images/org/health/articles/4471-pneumonia-01)

# What causes pneumonia?

Bacterial, viral, and fungal infections can cause pneumonia.

**Bacteria** are the most common cause. Bacterial pneumonia can occur on its own. It can also develop after you've had certain viral infections such as a cold or the flu. Several different types of bacteria can cause pneumonia, including:

* Streptococcus pneumoniae
* Legionella pneumophila; this type of pneumonia is often called Legionnaires' disease
* Mycoplasma pneumoniae
* Chlamydia pneumoniae
* Haemophilus influenzae type b (Hib)

  
**Viruses** that infect the respiratory tract may cause pneumonia. Viral pneumonia is often mild and goes away on its own within a few weeks. But sometimes it is serious enough that you need to get treatment in a hospital. If you have viral pneumonia, you are at risk of also getting bacterial pneumonia. The different viruses that can cause pneumonia include:

* Respiratory syncytial virus (RSV)
* Some common cold and flu viruses
* SARS-CoV-2, the virus that causes COVID-19

**Fungal** pneumonia is more common in people who have chronic health problems or weakened immune systems. Some of the types include:

* Pneumocystis pneumonia (PCP)
* Valley fever (Coccidioidomycosis)
* Histoplasmosis
* Cryptococcal pneumonia

Aspiration can also cause pneumonia. It can happen when you aspirate (accidentally breathe in) food or fluid into the lungs. If you are not able to cough out the food or fluid, it may cause an infection.

```python
# Pneumonia Hybrid Segmentation-Classification Pipeline

import os
import random
import copy
import numpy as np
import pandas as pd
from PIL import Image
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_curve, auc, confusion_matrix
)
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
import matplotlib.pyplot as plt

# -----------------------
# Dataset_paths 
# -----------------------
dataset_paths = {
    "dataset1_root": "/kaggle/input/chest-xray-pneumonia/chest_xray",  # change as needed
    "provided_test_path": "/kaggle/input/chest-xray-pneumonia/chest_xray/test",
    "artifacts_root": "/kaggle/working/artifacts"
}
segmentation_datasets = [  # List segmentation datasets here if available.
    # "/kaggle/input/shenzhen", "/kaggle/input/montgomery"
]
use_segmentation = True if len(segmentation_datasets) > 0 else False
dataset_paths['masks_root'] = None  # e.g. "/kaggle/input/pneumonia-masks/"

config = {
    "image_size": 224,
    "batch_size": 16,
    "num_epochs_seg": 30,
    "num_epochs_cls": 10,
    "lr": 1e-4,
    "weight_decay": 1e-5,
    "seed": 42,
    "num_workers": 2,
    "device": 'cuda' if torch.cuda.is_available() else 'cpu',
    "early_stopping_patience": 6,
    "min_val_size": 200,
}

print("Device ->", config['device'])

# Seed everything
def seed_everything(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

seed_everything(config['seed'])

# Dataset classes
class ChestXRayDataset(Dataset):
    def __init__(self, df: pd.DataFrame, image_root: str, masks_root: str = None,
                 predicted_mask_dir: str = None, image_size=224, augment=False):
        self.df = df.reset_index(drop=True)
        self.image_root = image_root
        self.masks_root = masks_root
        self.predicted_mask_dir = predicted_mask_dir
        self.image_size = image_size
        self.augment = augment
        self.base_transforms = transforms.Compose([
            transforms.Resize((image_size, image_size)),
            transforms.ToTensor(),
            transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
        ])
        self.aug_transforms = transforms.Compose([
            transforms.RandomHorizontalFlip(),
            transforms.RandomRotation(8),
            transforms.ColorJitter(brightness=0.08, contrast=0.08),
        ])
    def __len__(self):
        return len(self.df)
    def _load_image(self, rel_path):
        full = os.path.join(self.image_root, rel_path)
        if not os.path.exists(full):
            if os.path.exists(rel_path):
                full = rel_path
            else:
                raise FileNotFoundError(f"{full} not found")
        img = Image.open(full).convert('RGB')
        return img
    def _load_mask(self, rel_path):
        if self.predicted_mask_dir:
            pred = os.path.join(self.predicted_mask_dir, rel_path.replace(os.sep, "_") + ".png")
            if os.path.exists(pred):
                return Image.open(pred).convert('L')
        if self.masks_root:
            mask_path = os.path.join(self.masks_root, rel_path)
            if os.path.exists(mask_path):
                return Image.open(mask_path).convert('L')
        return None
    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        img = self._load_image(row['rel_path'])
        mask = self._load_mask(row.get('mask_rel_path', row['rel_path']))
        if self.augment:
            img = self.aug_transforms(img)
            if mask is not None:
                mask = self.aug_transforms(mask)
        img_t = self.base_transforms(img)
        if mask is not None:
            mask = mask.resize((self.image_size, self.image_size))
            mask_t = transforms.ToTensor()(mask)
            mask_t = (mask_t > 0.5).float()
        else:
            mask_t = torch.zeros((1, self.image_size, self.image_size), dtype=torch.float)
        label = torch.tensor(int(row['label']), dtype=torch.long)
        return img_t, mask_t, label, row['rel_path']

class SegmentationDataset(Dataset):
    def __init__(self, df: pd.DataFrame, image_root: str, mask_root: str,
                 image_size=224, augment=False):
        self.df = df.reset_index(drop=True)
        self.image_root = image_root
        self.mask_root = mask_root
        self.image_size = image_size
        self.augment = augment
        self.base_transforms = transforms.Compose([
            transforms.Resize((image_size, image_size)),
            transforms.ToTensor(),
        ])
        self.aug_transforms = transforms.Compose([
            transforms.RandomHorizontalFlip(),
            transforms.RandomRotation(8),
        ])
    def __len__(self):
        return len(self.df)
    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        img = Image.open(os.path.join(self.image_root, row['rel_path'])).convert('RGB')
        mask = Image.open(os.path.join(self.mask_root, row['mask_rel_path'])).convert('L')
        if self.augment:
            img = self.aug_transforms(img)
            mask = self.aug_transforms(mask)
        img_t = self.base_transforms(img)
        mask = mask.resize((self.image_size, self.image_size))
        mask_t = transforms.ToTensor()(mask)
        mask_t = (mask_t > 0.5).float()
        return img_t, mask_t, row['rel_path']

# -------------------------------------
# Helper functions to build dataframes
#--------------------------------------
def build_dataframe_from_kaggle_chestx(root_path):
    rows = []
    for split in ['train','val','test']:
        split_path = os.path.join(root_path, split)
        if not os.path.exists(split_path):
            continue
        for label_name in os.listdir(split_path):
            label_dir = os.path.join(split_path, label_name)
            if not os.path.isdir(label_dir):
                continue
            label_val = 0 if label_name.lower().startswith('normal') else 1
            for fname in sorted(os.listdir(label_dir)):
                if fname.lower().endswith(('.png','.jpg','.jpeg')):
                    rel_path = os.path.join(split, label_name, fname)
                    rows.append({'rel_path': rel_path, 'label': label_val, 'mask_rel_path': None})
    return pd.DataFrame(rows)

def build_segmentation_df_from_folder(images_root, masks_root):
    rows = []
    for fname in sorted(os.listdir(images_root)):
        if fname.lower().endswith(('.png','.jpg','.jpeg')):
            rows.append({'rel_path': fname, 'mask_rel_path': fname})
    return pd.DataFrame(rows)

# Build main dataframe
if os.path.exists(dataset_paths['dataset1_root']):
    df1 = build_dataframe_from_kaggle_chestx(dataset_paths['dataset1_root'])
else:
    df1 = pd.DataFrame(columns=['rel_path','label','mask_rel_path'])
print("Total samples:", len(df1))

# Ensure stable train/val/test splits
def ensure_valid_splits(df):
    if any([p.startswith('train'+os.sep) for p in df['rel_path'].tolist()]):
        train_df = df[df['rel_path'].str.startswith('train'+os.sep)].reset_index(drop=True)
        val_df = df[df['rel_path'].str.startswith('val'+os.sep)].reset_index(drop=True)
        test_df = df[df['rel_path'].str.startswith('test'+os.sep)].reset_index(drop=True)
    else:
        train_df, temp_df = train_test_split(df, test_size=0.2, stratify=df['label'], random_state=config['seed'])
        val_df, test_df = train_test_split(temp_df, test_size=0.5, stratify=temp_df['label'], random_state=config['seed'])
    total = len(train_df) + len(val_df) + len(test_df)
    if len(val_df) < max(int(0.08*total), config['min_val_size']):
        print(f"Validation set too small ({len(val_df)}). Re-splitting train for val set.")
        train_df, val_df = train_test_split(train_df, test_size=0.15, stratify=train_df['label'], random_state=config['seed'])
        train_df = train_df.reset_index(drop=True)
        val_df = val_df.reset_index(drop=True)
    return train_df, val_df, test_df

train_df, val_df, test_df = ensure_valid_splits(df1)
print(f"Train: {len(train_df)}, Val: {len(val_df)}, Test: {len(test_df)}")
print("Class balance in train:", train_df['label'].value_counts().to_dict())

# Create classification dataloaders
def create_classification_dataloaders(train_df, val_df, test_df, image_root, masks_root=None, predicted_mask_dir=None):
    classes = np.unique(train_df['label'])
    class_weights = compute_class_weight('balanced', classes=classes, y=train_df['label'].to_numpy())
    class_weights_tensor = torch.tensor(class_weights, dtype=torch.float).to(config['device'])
    print("Class weights:", dict(zip(classes, class_weights.tolist())))
    train_dataset = ChestXRayDataset(train_df, image_root=image_root, masks_root=masks_root,
                                     predicted_mask_dir=predicted_mask_dir, image_size=config['image_size'], augment=True)
    val_dataset = ChestXRayDataset(val_df, image_root=image_root, masks_root=masks_root,
                                   predicted_mask_dir=predicted_mask_dir, image_size=config['image_size'], augment=False)
    test_dataset = ChestXRayDataset(test_df, image_root=image_root, masks_root=masks_root,
                                    predicted_mask_dir=predicted_mask_dir, image_size=config['image_size'], augment=False)
    train_loader = DataLoader(train_dataset, batch_size=config['batch_size'], shuffle=True, num_workers=config['num_workers'])
    val_loader = DataLoader(val_dataset, batch_size=config['batch_size'], shuffle=False, num_workers=config['num_workers'])
    test_loader = DataLoader(test_dataset, batch_size=config['batch_size'], shuffle=False, num_workers=config['num_workers'])
    return train_loader, val_loader, test_loader, class_weights_tensor

train_loader, val_loader, test_loader, class_weights = create_classification_dataloaders(
    train_df, val_df, test_df, image_root=dataset_paths['dataset1_root'], masks_root=dataset_paths['masks_root'])

#====================================
# Define UNet and DenseNet models
#====================================

class UNetEncoderBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, padding=1),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, padding=1),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True),
        )
    def forward(self, x):
        return self.conv(x)

class UNet(nn.Module):
    def __init__(self, in_channels=3, out_channels=1, features=[32, 64, 128, 256]):
        super().__init__()
        self.downs = nn.ModuleList()
        self.ups = nn.ModuleList()
        self.pool = nn.MaxPool2d(2)
        for feat in features:
            self.downs.append(UNetEncoderBlock(in_channels, feat))
            in_channels = feat
        self.bottleneck = UNetEncoderBlock(features[-1], features[-1]*2)
        for feat in reversed(features):
            self.ups.append(nn.ConvTranspose2d(feat*2, feat, 2, 2))
            self.ups.append(UNetEncoderBlock(feat*2, feat))
        self.final_conv = nn.Conv2d(features[0], out_channels, kernel_size=1)
    def forward(self, x):
        skip_connections = []
        out = x
        for down in self.downs:
            out = down(out)
            skip_connections.append(out)
            out = self.pool(out)
        out = self.bottleneck(out)
        skip_connections = skip_connections[::-1]
        for idx in range(0, len(self.ups), 2):
            out = self.ups[idx](out)
            skip = skip_connections[idx//2]
            if out.shape != skip.shape:
                out = F.interpolate(out, size=skip.shape[2:])
            out = torch.cat((skip, out), dim=1)
            out = self.ups[idx+1](out)
        return torch.sigmoid(self.final_conv(out))

class DenseNetClassifier(nn.Module):
    def __init__(self, pretrained=True, num_classes=2):
        super().__init__()
        try:
            from torchvision.models import DenseNet121_Weights
            self.backbone = models.densenet121(weights=DenseNet121_Weights.IMAGENET1K_V1)
        except Exception:
            self.backbone = models.densenet121(pretrained=pretrained)
        num_ftrs = self.backbone.classifier.in_features
        self.backbone.classifier = nn.Identity()
        self.classifier = nn.Sequential(
            nn.AdaptiveAvgPool2d((1,1)),
            nn.Flatten(),
            nn.Dropout(0.5),
            nn.Linear(num_ftrs, num_classes)
        )
    def forward(self, x):
        features = self.backbone.features(x)
        out = self.classifier(features)
        return out

class HybridModel(nn.Module):
    def __init__(self, seg_model, cls_model, use_mask_multiply=True):
        super().__init__()
        self.seg_model = seg_model
        self.cls_model = cls_model
        self.use_mask_multiply = use_mask_multiply
    def forward(self, x):
        mask = self.seg_model(x)
        if self.use_mask_multiply:
            x_masked = x * mask
        else:
            x_masked = torch.cat([x, mask], dim=1)
        logits = self.cls_model(x_masked)
        return logits, mask

# Initialize models
seg_model = UNet(in_channels=3, out_channels=1).to(config['device'])
cls_model = DenseNetClassifier(pretrained=True, num_classes=2).to(config['device'])
hybrid_model = HybridModel(seg_model, cls_model, use_mask_multiply=True).to(config['device'])

# Loss functions
bce_loss = nn.BCELoss()
def dice_loss(pred, target, eps=1e-6):
    pred = pred.contiguous()
    target = target.contiguous()
    intersection = (pred * target).sum(dim=(2,3))
    union = pred.sum(dim=(2,3)) + target.sum(dim=(2,3))
    dice = (2.0 * intersection + eps) / (union + eps)
    return 1 - dice.mean()
def seg_loss_fn(pred, target):
    return bce_loss(pred, target) + dice_loss(pred, target)

# Optimizers, schedulers, amp scaler
optimizer_seg = torch.optim.Adam(seg_model.parameters(), lr=config['lr'], weight_decay=config['weight_decay'])
optimizer_cls = torch.optim.Adam(list(seg_model.parameters()) + list(cls_model.parameters()), lr=config['lr'], weight_decay=config['weight_decay'])
scheduler_seg = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer_seg, mode='min', factor=0.5, patience=3)
scheduler_cls = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer_cls, mode='min', factor=0.5, patience=3)
scaler = torch.amp.GradScaler('cuda')

# Early Stopping class
class EarlyStopping:
    def __init__(self, patience=6, mode='max', min_delta=0.0):
        self.patience = patience
        self.mode = mode
        self.min_delta = min_delta
        self.best = None
        self.counter = 0
    def step(self, value):
        if self.best is None:
            self.best = value
            self.counter = 0
            return False
        improved = ((value - self.best) > self.min_delta) if self.mode == 'max' else ((self.best - value) > self.min_delta)
        if improved:
            self.best = value
            self.counter = 0
            return False
        else:
            self.counter += 1
            return self.counter >= self.patience

# Training and evaluation functions
def train_segmentation(model, train_loader, val_loader, optimizer, scheduler, scaler, num_epochs=10, device='cuda'):
    model.to(device)
    best_wts = copy.deepcopy(model.state_dict())
    best_loss = float('inf')
    history = {'train_loss': [], 'val_loss': []}
    early_stop = EarlyStopping(patience=config['early_stopping_patience'], mode='min')
    for epoch in range(num_epochs):
        model.train()
        running = 0.0
        for imgs, masks, *_ in train_loader:
            imgs, masks = imgs.to(device), masks.to(device)
            optimizer.zero_grad()
            with torch.amp.autocast('cuda'):
                preds = model(imgs)
                loss = seg_loss_fn(preds, masks)
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
            running += loss.item() * imgs.size(0)
        train_loss = running / len(train_loader.dataset)
        history['train_loss'].append(train_loss)
        model.eval()
        val_running = 0.0
        with torch.no_grad():
            for imgs, masks, *_ in val_loader:
                imgs, masks = imgs.to(device), masks.to(device)
                with torch.amp.autocast('cuda'):
                    preds = model(imgs)
                    loss = seg_loss_fn(preds, masks)
                val_running += loss.item() * imgs.size(0)
        val_loss = val_running / len(val_loader.dataset)
        history['val_loss'].append(val_loss)
        scheduler.step(val_loss)
        print(f"[Seg] Epoch {epoch+1}/{num_epochs} train_loss={train_loss:.4f} val_loss={val_loss:.4f}")
        if val_loss < best_loss:
            best_loss = val_loss
            best_wts = copy.deepcopy(model.state_dict())
        if early_stop.step(-val_loss):
            print("Seg Early stopping")
            break
    model.load_state_dict(best_wts)
    return model, history

def train_classification(model, train_loader, val_loader, optimizer, scheduler, criterion, scaler, num_epochs=10, device='cuda'):
    model.to(device)
    best_wts = copy.deepcopy(model.state_dict())
    best_auc = 0.0
    history = {'train_loss': [], 'val_loss': [], 'val_auc': []}
    early_stop = EarlyStopping(patience=config['early_stopping_patience'], mode='max')
    for epoch in range(num_epochs):
        model.train()
        running = 0.0
        for imgs, masks, labels, _ in train_loader:
            imgs, labels = imgs.to(device), labels.to(device)
            optimizer.zero_grad()
            with torch.amp.autocast('cuda'):
                logits, _ = model(imgs)
                loss = criterion(logits, labels)
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
            running += loss.item() * imgs.size(0)
        train_loss = running / len(train_loader.dataset)
        history['train_loss'].append(train_loss)
        model.eval()
        val_running = 0.0
        all_probs, all_labels = [], []
        with torch.no_grad():
            for imgs, masks, labels, _ in val_loader:
                imgs, labels = imgs.to(device), labels.to(device)
                with torch.amp.autocast('cuda'):
                    logits, _ = model(imgs)
                    loss = criterion(logits, labels)
                val_running += loss.item() * imgs.size(0)
                probs = torch.softmax(logits, dim=1)[:,1].cpu().numpy()
                all_probs.extend(probs.tolist())
                all_labels.extend(labels.cpu().numpy().tolist())
        val_loss = val_running / len(val_loader.dataset)
        fpr, tpr, _ = roc_curve(all_labels, all_probs) if len(np.unique(all_labels)) > 1 else ([0,1],[0,1],None)
        val_auc = auc(fpr, tpr)
        history['val_loss'].append(val_loss)
        history['val_auc'].append(val_auc)
        scheduler.step(val_loss)
        print(f"[Cls] Epoch {epoch+1}/{num_epochs} train_loss={train_loss:.4f} val_loss={val_loss:.4f} val_auc={val_auc:.4f}")
        if val_auc > best_auc:
            best_auc = val_auc
            best_wts = copy.deepcopy(model.state_dict())
        if early_stop.step(val_auc):
            print("Cls Early stopping")
            break
    model.load_state_dict(best_wts)
    return model, history

def train_baseline(baseline, train_loader, val_loader, optimizer, criterion, scaler, num_epochs=3, device='cuda'):
    baseline.to(device)
    best_wts = copy.deepcopy(baseline.state_dict())
    best_auc = 0.0
    history = {'train_loss':[], 'val_loss':[], 'val_auc':[]}
    early_stop = EarlyStopping(patience=config['early_stopping_patience'], mode='max')
    for epoch in range(num_epochs):
        baseline.train()
        running = 0.0
        for imgs, masks, labels, _ in train_loader:
            imgs, labels = imgs.to(device), labels.to(device)
            optimizer.zero_grad()
            with torch.amp.autocast('cuda'):
                logits = baseline(imgs)
                loss = criterion(logits, labels)
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
            running += loss.item() * imgs.size(0)
        train_loss = running / len(train_loader.dataset)
        history['train_loss'].append(train_loss)
        baseline.eval()
        val_running = 0.0
        all_probs, all_labels = [], []
        with torch.no_grad():
            for imgs, masks, labels, _ in val_loader:
                imgs, labels = imgs.to(device), labels.to(device)
                with torch.amp.autocast('cuda'):
                    logits = baseline(imgs)
                    loss = criterion(logits, labels)
                val_running += loss.item() * imgs.size(0)
                probs = torch.softmax(logits, dim=1)[:,1].cpu().numpy()
                all_probs.extend(probs.tolist())
                all_labels.extend(labels.cpu().numpy().tolist())
        val_loss = val_running / len(val_loader.dataset)
        fpr, tpr, _ = roc_curve(all_labels, all_probs) if len(np.unique(all_labels))>1 else ([0,1],[0,1],None)
        val_auc = auc(fpr,tpr)
        history['val_loss'].append(val_loss)
        history['val_auc'].append(val_auc)
        print(f"[Baseline] Epoch {epoch+1}/{num_epochs} train_loss={train_loss:.4f} val_loss={val_loss:.4f} val_auc={val_auc:.4f}")
        if val_auc > best_auc:
            best_auc = val_auc
            best_wts = copy.deepcopy(baseline.state_dict())
        if early_stop.step(val_auc):
            print("Baseline Early stopping")
            break
    baseline.load_state_dict(best_wts)
    return baseline, history

def evaluate_model(model, loader, device='cuda'):
    model.eval()
    all_probs, all_preds, all_labels = [], [], []
    with torch.no_grad():
        for imgs, masks, labels, paths in loader:
            imgs, labels = imgs.to(device), labels.to(device)
            logits, _ = model(imgs)
            probs = torch.softmax(logits, dim=1)[:,1]
            preds = (probs>0.5).long()
            all_probs.extend(probs.cpu().numpy().tolist())
            all_preds.extend(preds.cpu().numpy().tolist())
            all_labels.extend(labels.cpu().numpy().tolist())
    acc = accuracy_score(all_labels, all_preds)
    prec = precision_score(all_labels, all_preds, zero_division=0)
    rec = recall_score(all_labels, all_preds, zero_division=0)
    f1 = f1_score(all_labels, all_preds, zero_division=0)
    fpr, tpr, _ = roc_curve(all_labels, all_probs) if len(np.unique(all_labels))>1 else ([0,1],[0,1],None)
    roc_auc = auc(fpr, tpr)
    cm = confusion_matrix(all_labels, all_preds)
    return {'accuracy':acc,'precision':prec,'recall':rec,'f1':f1,'auc':roc_auc,'fpr':fpr,'tpr':tpr,'cm':cm,'probs':all_probs,'labels':all_labels}

def evaluate_baseline_model(model, loader, device='cuda'):
    model.eval()
    all_probs, all_preds, all_labels = [], [], []
    with torch.no_grad():
        for imgs, masks, labels, paths in loader:
            imgs, labels = imgs.to(device), labels.to(device)
            logits = model(imgs)
            probs = torch.softmax(logits, dim=1)[:,1]
            preds = (probs>0.5).long()
            all_probs.extend(probs.cpu().numpy().tolist())
            all_preds.extend(preds.cpu().numpy().tolist())
            all_labels.extend(labels.cpu().numpy().tolist())
    acc = accuracy_score(all_labels, all_preds)
    prec = precision_score(all_labels, all_preds, zero_division=0)
    rec = recall_score(all_labels, all_preds, zero_division=0)
    f1 = f1_score(all_labels, all_preds, zero_division=0)
    fpr, tpr, _ = roc_curve(all_labels, all_probs) if len(np.unique(all_labels))>1 else ([0,1],[0,1],None)
    roc_auc = auc(fpr, tpr)
    cm = confusion_matrix(all_labels, all_preds)
    return {'accuracy':acc,'precision':prec,'recall':rec,'f1':f1,'auc':roc_auc,'fpr':fpr,'tpr':tpr,'cm':cm}

#================================================
# Grad-CAM class omitted here for brevity; can be added as before
#================================================

def generate_and_cache_masks(seg_model, df, image_root, save_dir):
    os.makedirs(save_dir, exist_ok=True)
    seg_model.eval()
    with torch.no_grad():
        for _, row in df.reset_index(drop=True).iterrows():
            rel = row['rel_path']
            try:
                img = Image.open(os.path.join(image_root, rel)).convert('RGB')
            except:
                img = Image.open(rel).convert('RGB')
            img_resized = transforms.Resize((config['image_size'], config['image_size']))(img)
            img_t = transforms.ToTensor()(img_resized)
            img_t = transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])(img_t).unsqueeze(0).to(config['device'])
            with torch.amp.autocast('cuda'):
                pred = seg_model(img_t)
            pred_np = (pred.squeeze().cpu().numpy() > 0.5).astype(np.uint8)*255
            save_name = rel.replace(os.sep, "_") + ".png"
            Image.fromarray(pred_np).save(os.path.join(save_dir, save_name))
    print("Predicted masks saved to:", save_dir)
    return save_dir

# Main pipeline execution starts here
art_root = dataset_paths.get('artifacts_root', './artifacts')
os.makedirs(art_root, exist_ok=True)

# 1) Train segmentation if datasets provided (optional)
if use_segmentation and seg_model is not None:
    print("Training segmentation on provided segmentation datasets...")
    seg_img_paths, seg_mask_paths = [], []
    for seg_root in segmentation_datasets:
        imgs = os.path.join(seg_root, 'images')
        masks = os.path.join(seg_root, 'masks')
        if os.path.exists(imgs) and os.path.exists(masks):
            for fname in sorted(os.listdir(imgs)):
                if fname.lower().endswith(('.jpg','.png','.jpeg')):
                    seg_img_paths.append(os.path.join(imgs, fname))
                    seg_mask_paths.append(os.path.join(masks, fname))
    if len(seg_img_paths) == 0:
        print("No segmentation files found; skipping segmentation training.")
        use_segmentation = False
    else:
        seg_df = pd.DataFrame({'rel_path': seg_img_paths, 'mask_rel_path': seg_mask_paths})
        s_train, s_val = train_test_split(seg_df, test_size=0.15, random_state=config['seed'])
        seg_train_ds = SegmentationDataset(s_train, image_root="", mask_root="", image_size=config['image_size'], augment=True)
        seg_val_ds = SegmentationDataset(s_val, image_root="", mask_root="", image_size=config['image_size'], augment=False)
        seg_train_loader = DataLoader(seg_train_ds, batch_size=max(4, config['batch_size']//4), shuffle=True, num_workers=0)
        seg_val_loader = DataLoader(seg_val_ds, batch_size=max(4, config['batch_size']//4), shuffle=False, num_workers=0)
        seg_model, seg_hist = train_segmentation(seg_model, seg_train_loader, seg_val_loader,
                                                 optimizer_seg, scheduler_seg, scaler,
                                                 num_epochs=config['num_epochs_seg'], device=config['device'])
        torch.save(seg_model.state_dict(), os.path.join(art_root, 'seg_model.pth'))

# 2) Generate predicted masks for pneumonia dataset (if segmentation used)
pred_mask_dir = None
if use_segmentation:
    print("Generating predicted lung masks for pneumonia dataset (take time)...")
    pred_mask_dir = os.path.join(art_root, 'predicted_masks')
    generate_and_cache_masks(seg_model, pd.concat([train_df, val_df, test_df], ignore_index=True),
                             image_root=dataset_paths['dataset1_root'], save_dir=pred_mask_dir)

# 3) Prepare classification dataloaders with predicted masks if available
train_loader, val_loader, test_loader, class_weights = create_classification_dataloaders(
    train_df, val_df, test_df, image_root=dataset_paths['dataset1_root'],
    masks_root=dataset_paths['masks_root'], predicted_mask_dir=pred_mask_dir)

cls_criterion = nn.CrossEntropyLoss(weight=class_weights)

# Freeze seg_model for classifier training
for p in seg_model.parameters():
    p.requires_grad = False

# Train hybrid classifier
hybrid_model = HybridModel(seg_model, cls_model, use_mask_multiply=True).to(config['device'])
optimizer_cls = torch.optim.Adam(filter(lambda p: p.requires_grad, hybrid_model.parameters()),
                                 lr=config['lr'], weight_decay=config['weight_decay'])
hybrid_model, cls_history = train_classification(hybrid_model, train_loader, val_loader,
                                                optimizer_cls, scheduler_cls, cls_criterion, scaler,
                                                num_epochs=config['num_epochs_cls'], device=config['device'])
torch.save(hybrid_model.state_dict(), os.path.join(art_root, 'hybrid_model.pth'))

# Evaluate hybrid
hybrid_results = evaluate_model(hybrid_model, test_loader, device=config['device'])
print("Hybrid Test results:", {k: hybrid_results[k] for k in ['accuracy','precision','recall','f1','auc']})
plt.figure(figsize=(6,6))
plt.plot(hybrid_results['fpr'], hybrid_results['tpr'], label=f"AUC={hybrid_results['auc']:.4f}")
plt.plot([0,1],[0,1],'--')
plt.legend()
plt.title("Hybrid ROC")
plt.show()
print("Confusion Matrix:\n", hybrid_results['cm'])

# 4) Baseline DenseNet (no segmentation)
print("Training baseline DenseNet (no segmentation)...")
baseline = DenseNetClassifier(pretrained=True, num_classes=2).to(config['device'])
optimizer_baseline = torch.optim.Adam(baseline.parameters(), lr=config['lr'], weight_decay=config['weight_decay'])
baseline, base_hist = train_baseline(baseline, train_loader, val_loader, optimizer_baseline, cls_criterion, scaler,
                                     num_epochs=10, device=config['device'])
torch.save(baseline.state_dict(), os.path.join(art_root, 'baseline_model.pth'))

baseline_results = evaluate_baseline_model(baseline, test_loader, device=config['device'])
print("Baseline test results:", {k: baseline_results[k] for k in ['accuracy','precision','recall','f1','auc']})
plt.figure(figsize=(6,6))
plt.plot(baseline_results['fpr'], baseline_results['tpr'], label=f"AUC={baseline_results['auc']:.4f}")
plt.plot([0,1],[0,1],'--')
plt.title("Baseline ROC")
plt.legend()
plt.show()
print("Baseline Confusion Matrix:\n", baseline_results['cm'])

# 5) Save report
import json
report = {
    'hybrid': {k: float(hybrid_results[k]) for k in ['accuracy','precision','recall','f1','auc']},
    'baseline': {k: float(baseline_results[k]) for k in ['accuracy','precision','recall','f1','auc']}
}
with open(os.path.join(art_root, 'report.json'), 'w') as f:
    json.dump(report, f, indent=2)
print("Saved report to", os.path.join(art_root, 'report.json'))

# Grad-CAM visualization code can be added here as needed.

print("Pipeline finished. Check artifacts in:", art_root)
```

# UNet (Segmentation Backbone):

The UNet model is a convolutional neural network architecture designed for precise pixel-level segmentation. It consists of a contracting path that captures context via repeated convolution and pooling, followed by an expanding path that enables precise localization through upsampling and skip connections. In your pipeline, UNet is implemented with four encoder blocks (featuring two 33 convolutions, batch normalization, and ReLU), a bottleneck layer, and a corresponding decoder path using transposed convolutions and concatenation of encoder feature maps. The final 11 convolution produces a single-channel lung mask, which is passed through a sigmoid to yield a probability map.

# DenseNet121 (Classification Backbone):

DenseNet121 is a densely connected convolutional network that promotes feature reuse and mitigates the vanishing-gradient problem. Each layer receives inputs from all preceding layers via feature-map concatenation, improving gradient flow and parameter efficiency. In your code, the pretrained DenseNet121 from torchvision is modified by removing its default classifier and replacing it with an adaptive average pooling layer, dropout, and a linear layer to output two logits (Normal vs. Pneumonia). The DenseNet encoder extracts rich image representations, which are then classified into the two pneumonia categories.

# Pipeline Architecture and Methodology

The provided code implements a two-stage hybrid approach that first performs lung segmentation using U-Net, then applies the segmentation masks to enhance pneumonia classification using DenseNet. This methodology addresses a critical limitation in traditional pneumonia detection systems by explicitly focusing the classification model on relevant lung regions while filtering out background noise and irrelevant anatomical structures

# T-SNE Feature Embedding Extraction and Visualization


This snippet first extracts deep features from the DenseNet backbone for each test image and then projects those high-dimensional embeddings down to two dimensions using t-SNE. The 2D scatter plot highlights how Normal vs. Pneumonia cases cluster in feature space.


```python
from sklearn.manifold import TSNE

def extract_features(model, loader, device='cuda'):
    model.eval()
    features_list, labels_list = [], []
    with torch.no_grad():
        for imgs, masks, labels, _ in loader:
            imgs = imgs.to(device)
            # Extract features from DenseNet backbone (before classifier)
            features = model.cls_model.backbone.features(imgs)
            pooled = F.adaptive_avg_pool2d(features, (1,1)).reshape(imgs.size(0), -1)
            features_list.append(pooled.cpu())
            labels_list.append(labels)
    features_tensor = torch.cat(features_list, dim=0)
    labels_tensor = torch.cat(labels_list, dim=0)
    return features_tensor.numpy(), labels_tensor.numpy()

def plot_tsne(features, labels, perplexity=30, n_iter=1000):
    tsne = TSNE(n_components=2, perplexity=perplexity, n_iter=n_iter, random_state=42)
    tsne_results = tsne.fit_transform(features)
    plt.figure(figsize=(8,6))
    scatter = plt.scatter(tsne_results[:,0], tsne_results[:,1], c=labels, cmap='coolwarm', alpha=0.7)
    plt.legend(handles=scatter.legend_elements()[0], labels=['Normal', 'Pneumonia'])
    plt.title("t-SNE Visualization of Feature Embeddings")
    plt.xlabel("t-SNE dim 1")
    plt.ylabel("t-SNE dim 2")
    plt.grid(True)
    plt.show()

# Usage example:
features, labels = extract_features(hybrid_model, test_loader, device=config['device'])
plot_tsne(features, labels)
```

# Grad-CAM (Gradient-weighted Class Activation Mapping) 

Grad CAM is a powerful technique used to generate heatmaps that highlight the important regions in chest X-ray images contributing to a deep learning model's decision, such as detecting pneumonia. It works by computing the gradients of the predicted class score with respect to feature maps in the last convolutional layers, then producing a localization map that indicates which parts of the X-ray image most influenced the class prediction.

```python
import torch
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
from torchvision import transforms

class GradCAM:
    def __init__(self, model, target_layer):
        self.model = model
        self.target_layer = target_layer
        self.model.eval()
        self.activations = None
        self.gradients = None
        self.hook_handles = []
        self._register_hooks()

    def _register_hooks(self):
        def forward_hook(module, input, output):
            self.activations = output.detach()

        def backward_hook(module, grad_input, grad_output):
            self.gradients = grad_output[0].detach()

        # Hook the specified layer
        self.hook_handles.append(self.target_layer.register_forward_hook(forward_hook))
        self.hook_handles.append(self.target_layer.register_backward_hook(backward_hook))

    def __del__(self):
        # Remove hooks
        for handle in self.hook_handles:
            handle.remove()

    def generate(self, input_tensor, target_class=None):
        self.model.zero_grad()
        output = self.model(input_tensor)
        if target_class is None:
            target_class = output.argmax(dim=1).item()
        loss = output[:, target_class].sum()
        loss.backward()

        # Get pooled gradients
        pooled_gradients = torch.mean(self.gradients, dim=(2, 3), keepdim=True)
        # Weight activations
        activations = self.activations
        weighted_activations = pooled_gradients * activations
        heatmap = weighted_activations.sum(dim=1).squeeze()

        # ReLU and normalize
        heatmap = torch.clamp(heatmap, min=0)
        heatmap = heatmap / (heatmap.max() + 1e-8)
        heatmap = heatmap.cpu().numpy()
        return heatmap

def visualize_gradcam(image_tensor, model, config, class_names, target_class=None, alpha=0.4):
    # Find the last convolutional layer of DenseNet
    target_layer = model.backbone.features.norm5  # For DenseNet-121

    gradcam = GradCAM(model, target_layer)
    with torch.no_grad():
        input_tensor = image_tensor.unsqueeze(0).to(config['device'])
    # Gradients needed, so not in torch.no_grad
    input_tensor.requires_grad = True
    heatmap = gradcam.generate(input_tensor, target_class)
    
    # Un-normalize image for display
    mean = torch.tensor([0.485, 0.456, 0.406]).view(3, 1, 1)
    std = torch.tensor([0.229, 0.224, 0.225]).view(3, 1, 1)
    img_disp = image_tensor.cpu() * std + mean
    img_disp = img_disp.clamp(0,1).permute(1,2,0).numpy()

    # Resize heatmap to match image
    heatmap_resized = np.uint8(255 * heatmap)
    heatmap_resized = np.stack([heatmap_resized]*3, axis=-1)
    heatmap_resized = transforms.ToPILImage()(heatmap_resized)
    heatmap_resized = heatmap_resized.resize((img_disp.shape[1], img_disp.shape[0]))
    heatmap_resized = np.array(heatmap_resized)[:,:,0] / 255.0  # single channel

    # Overlay heatmap on image
    plt.figure(figsize=(7,7))
    plt.imshow(img_disp)
    plt.imshow(heatmap_resized, cmap='jet', alpha=alpha)
    plt.axis('off')
    if target_class is not None and class_names:
        plt.title(f"Activation heatmap ({class_names[target_class]})")
    plt.show()

# ------- Usage Example -------

# Take image from test_loader
img_t, mask_t, label, path = next(iter(test_loader))
img_sample = img_t[0]  # first sample
label_sample = label[0].item() if torch.is_tensor(label) else label[0]

class_names = ['Normal', 'Pneumonia']

# For hybrid model: pass only classifier for Grad-CAM (DenseNet)
visualize_gradcam(img_sample, cls_model, config, class_names, target_class=label_sample, alpha=0.35)
```
